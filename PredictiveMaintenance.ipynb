```python
from sklearn.preprocessing import LabelBinarizer
labelbinarizer = LabelBinarizer()
encoded_results_1 = labelbinarizer.fit_transform(predictive_data["Failure Type"])
```

```python

encoded_results_1
```

```python
predictive_data_encoded_1 = pd.DataFrame(encoded_results_1,columns=labelbinarizer.classes_)
```

```python
encoded_results_2 = labelbinarizer.fit_transform(predictive_data["Type"])


predictive_data_encoded_2 = pd.DataFrame(encoded_results_2,columns=labelbinarizer.classes_)


predictive_data_1 = pd.concat([predictive_data, predictive_data_encoded_1, predictive_data_encoded_2], axis=1)


predictive_data_1
```
